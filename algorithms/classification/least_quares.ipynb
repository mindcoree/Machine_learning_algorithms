# –õ–∏–Ω–µ–π–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è: –æ—Ç—Å—Ç—É–ø –∏ —Ñ—É–Ω–∫—Ü–∏—è –æ—à–∏–±–∫–∏

## üìò –û–ø–∏—Å–∞–Ω–∏–µ

–í —ç—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ—Ç—Å—è **–ª–∏–Ω–µ–π–Ω–∞—è –º–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏**, –æ–±—É—á–∞–µ–º–∞—è –Ω–∞ –¥–≤—É–º–µ—Ä–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
–¶–µ–ª—å ‚Äî –Ω–∞–π—Ç–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞–∏–ª—É—á—à–∏–º –æ–±—Ä–∞–∑–æ–º —Ä–∞–∑–¥–µ–ª—è—é—Ç –æ–±—ä–µ–∫—Ç—ã –¥–≤—É—Ö –∫–ª–∞—Å—Å–æ–≤ $y_i \in \{-1, 1\}$.

---

## üî¢ –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

–ö–∞–∂–¥–æ–º—É –æ–±—ä–µ–∫—Ç—É —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –≤–µ–∫—Ç–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:

$$
x_i = (x_{i1}, x_{i2})
$$

–∏ —Ü–µ–ª–µ–≤–∞—è –º–µ—Ç–∫–∞ –∫–ª–∞—Å—Å–∞:

$$
y_i \in \{-1, 1\}.
$$

–î–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –≤ –º–∞—Ç—Ä–∏—Ü—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –µ–¥–∏–Ω–∏—á–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü, —á—Ç–æ–±—ã —É—á–µ—Å—Ç—å —Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω $w_0$:

$$
X =
\begin{bmatrix}
1 & x_{11} & x_{12} \\
1 & x_{21} & x_{22} \\
\vdots & \vdots & \vdots \\
1 & x_{n1} & x_{n2}
\end{bmatrix},
\quad
y =
\begin{bmatrix}
y_1 \\ y_2 \\ \vdots \\ y_n
\end{bmatrix}
$$

---

## ‚öôÔ∏è –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏ –Ω–∞—Ö–æ–¥—è—Ç—Å—è —Å –ø–æ–º–æ—â—å—é **–∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è** (–Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è):

$$
w = (X^T X)^{-1} X^T y
$$

–≥–¥–µ $w = [w_0, w_1, w_2]^T$ ‚Äî –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ –º–æ–¥–µ–ª–∏.

---

## üìà –û—Ç—Å—Ç—É–ø (margin)

–í–≤–µ–¥—ë–º –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ:

$$
M_i = y_i (w^T x_i)
$$

–í–µ–ª–∏—á–∏–Ω–∞ $M_i$ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ —É–≤–µ—Ä–µ–Ω–Ω–æ –∏ –ø—Ä–∞–≤–∏–ª—å–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω –æ–±—ä–µ–∫—Ç:

- $M_i > 0$ ‚Äî –æ–±—ä–µ–∫—Ç –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω **–ø—Ä–∞–≤–∏–ª—å–Ω–æ**,
- $M_i < 0$ ‚Äî –æ–±—ä–µ–∫—Ç –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω **–Ω–µ–≤–µ—Ä–Ω–æ**,
- —á–µ–º –±–æ–ª—å—à–µ $M_i$, —Ç–µ–º –≤—ã—à–µ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏.

---

## ‚ùå –ü–æ–¥—Å—á—ë—Ç –æ—à–∏–±–æ–∫ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—à–∏–±–æ–∫ –º–æ–∂–Ω–æ –≤—ã—Ä–∞–∑–∏—Ç—å —á–µ—Ä–µ–∑ –æ—Ç—Å—Ç—É–ø—ã:

$$
N_{\text{err}} = \sum_{i=1}^{n} [M_i < 0]
$$

–≥–¥–µ $[M_i < 0]$ ‚Äî –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è (—Ä–∞–≤–Ω–∞ 1, –µ—Å–ª–∏ –æ–±—ä–µ–∫—Ç –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω –Ω–µ–≤–µ—Ä–Ω–æ, –∏ 0 ‚Äî –µ—Å–ª–∏ –≤–µ—Ä–Ω–æ).

---

## ‚öñÔ∏è –§—É–Ω–∫—Ü–∏—è –æ—à–∏–±–∫–∏

–î–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å:

$$
Q = \sum_{i=1}^{n} (1 - M_i)^2
$$

–û–Ω–∞ —à—Ç—Ä–∞—Ñ—É–µ—Ç –∫–∞–∫ –æ—à–∏–±–∫–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏, —Ç–∞–∫ –∏ —Å–ª–∞–±—ã–µ (–Ω–µ—É–≤–µ—Ä–µ–Ω–Ω—ã–µ) —Ä–µ—à–µ–Ω–∏—è, –∫–æ–≥–¥–∞ $M_i$ –±–ª–∏–∑–∫–æ –∫ 1.

---

## üí° –ò—Ç–æ–≥

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –ø–æ–ª—É—á–∞–µ–º:

- **–≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ $w$** ‚Äî –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –ª–∏–Ω–µ–π–Ω–æ–π –º–æ–¥–µ–ª–∏,
- **–≤–µ–∫—Ç–æ—Ä –æ—Ç—Å—Ç—É–ø–æ–≤ $M$** ‚Äî –º–µ—Ä–∞ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏,
- **$N_{\text{err}}$** ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤,
- **$Q$** ‚Äî –æ–±—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è –æ—à–∏–±–∫–∏.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –ø—Ä–∏–º–µ—Ä –∏–ª–ª—é—Å—Ç—Ä–∏—Ä—É–µ—Ç –±–∞–∑–æ–≤—ã–µ –ø—Ä–∏–Ω—Ü–∏–ø—ã —Ä–∞–±–æ—Ç—ã **–ª–∏–Ω–µ–π–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞** ‚Äî –æ—Ç –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≤–µ—Å–æ–≤ –¥–æ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —á–µ—Ä–µ–∑ –æ—Ç—Å—Ç—É–ø—ã –∏ —Ñ—É–Ω–∫—Ü–∏—é –ø–æ—Ç–µ—Ä—å.
