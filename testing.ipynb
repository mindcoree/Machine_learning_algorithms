```python
eps = [i for i in range(10, 100, 5)]
min_samples = [i for i in range(1, 10)]
klstrs = []
n = []
params = []

for m_s in min_samples:
    for ep in eps:

        clustering = DBSCAN(eps=ep, min_samples=m_s, metric='euclidean')
        res = clustering.fit_predict(X)
        n.append(min(res))
        klstrs.append(np.unique(res).shape[0])
        params.append((m_s, ep))

for i, j in zip(klstrs, n):
    if i == 3 and j > -1:
        best_param = params[klstrs.index(i)]

ep = best_param[1]
m_s = best_param[0]
clustering = DBSCAN(eps=ep, min_samples=m_s, metric='euclidean')
res = clustering.fit_predict(X)
res.reshape(-1, 1).shape
```

```python
import numpy as np

# исходная функция, которую нужно аппроксимировать моделью a(x)
def func(x):
    return 0.5 * x**2 - 0.1 * 1/np.exp(-x) + 0.5 * np.cos(2*x) - 2.


# здесь объявляйте необходимые функции


coord_x = np.arange(-5.0, 5.0, 0.1).reshape(-1,1) # значения по оси абсцисс [-5; 5] с шагом 0.1
coord_y = func(coord_x) # значения функции по оси ординат
coord_y = np.array(coord_y).reshape(-1,1)
sz = len(coord_x)	# количество значений функций (точек)
eta = np.array([0.01, 0.001, 0.0001, 0.01, 0.01]).reshape(-1,1) # шаг обучения для каждого параметра w0, w1, w2, w3, w4
w = np.array([0., 0., 0., 0., 0.]).reshape(-1,1) # начальные значения параметров модели
N = 500 # число итераций алгоритма SGD
lm = 0.02 # значение параметра лямбда для вычисления скользящего экспоненциального среднего
Y = np.array(coord_y).reshape(-1,1)
X = np.hstack([
    np.ones_like(coord_x),
    coord_x,
    coord_x**2,
    np.cos(2*coord_x),
    np.sin(2*coord_x)  
])
np.random.seed(0) # генерация одинаковых последовательностей псевдослучайных чисел

# начальная среднеквадратичная ошибка
Qe = np.mean((X @ w - coord_y)**2)

# стохастический градиентный спуск
for _ in range(N):
    k = np.random.randint(0, sz-1)
    x_k = X[k].reshape(-1,1)   
    y_k = coord_y.reshape(-1,1) 
    print(x_k.shape, y_k.shape)      

    # градиент для одного наблюдения
    gradient_k = 2 * (x_k @ w - y_k) @ x_k  # (5, 1)

    # обновление весов (поэлементно с разными eta)
    w -= eta * gradient_k

    # обновление ошибки Qe
    loss_k = ((x_k @ w - y_k)**2).item() # скаляр
    Qe = lm * loss_k + (1 - lm) * Qe

Q = np.mean((X @ w - coord_y)**2)
```
